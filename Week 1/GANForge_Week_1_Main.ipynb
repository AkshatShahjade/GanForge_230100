```python
import matplotlib.pylab as plt
import cv2

from glob import glob
import numpy as np
import pandas as pd
```

```python
cat_photo = glob("./Input_data/Cats/cat.4001.jpg")[0]
```

```python
# Loading image by 2 methods:
cat_mpl = plt.imread(cat_photo) # Stored as ndarray[height,width,RGB]
cat_cv2 = cv2.imread(cat_photo) # Stored as ndarray[height,width,BRG]
```

```python
# converting cv2 BRG->RGB
cat_cv2_rgb = cv2.cvtColor(cat_cv2,cv2.COLOR_BGR2RGB)
cv2.cvtColor(cat_cv2, cv2.COLOR_)
```

```python
# Displaying Image
fig, axs = plt.subplots(1,3)
axs[0].imshow(cat_mpl)
axs[1].imshow(cat_cv2)
axs[2].imshow(cat_cv2_rgb)

axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')

plt.show()
```

```python
# Grayscaling
cat_cv2_grey = cv2.cvtColor(cat_cv2, cv2.COLOR_BGR2GRAY) # Takes a 3 stream image to single stream image
fig, ax = plt.subplots()
ax.imshow(cat_cv2_grey, cmap="Greys") # Gives the color to show the single stream of values
ax.axis('off')
plt.show()
```

```python
# Resizing
fig, axs = plt.subplots(1,3)
axs[0].imshow(cat_cv2)
cat_cv2_shrunk = cv2.resize(cat_cv2, None, fx=0.5,fy=0.5)
axs[1].imshow(cat_cv2_shrunk)
cat_cv2_resized = cv2.resize(cat_cv2, (100, 50))
axs[2].imshow(cat_cv2_resized)
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.imshow(cat_cv2)
fig2, ax2 = plt.subplots()
cat_cv2_resized = cv2.resize(cat_cv2, None, fx=5, fy=5, interpolation=cv2.INTER_CUBIC)
ax2.imshow(cat_cv2_resized)
plt.show()
```

```python
# Using any Kernel on an Image
exposure = 1000
kernel = exposure*np.array([[1,0,1],[0,-11,0],[1,0,1]])

filtered_img = cv2.filter2D(cat_cv2, -1, kernel)
fig, ax = plt.subplots()
ax.imshow(filtered_img)
plt.show()
```

```python
# Saving Images. We can save in png, or jpg, etc.
plt.imsave("single_channel_cat.png", cat_cv2_grey)
cv2.imwrite("cat_edge.jpg", filtered_img)
```
